```python
from vergelijkingen import mendelsohn
from matplotlib import pyplot as plt

dagen = []
volumes = []
t = 0
delta_t = 1
n = 10
start_vol = 10
c = 2
for i in range(n):
    if i == 0:
        t += delta_t
        delta_volume = mendelsohn(start_vol, c, delta_t)
        dagen.append(t)
        volumes.append(delta_volume)
    else:
        t += delta_t
        delta_volume = mendelsohn(volumes[i-1], c, delta_t)
        dagen.append(t)
        volumes.append(delta_volume)

print(dagen)
print(volumes)
#pd.DataFrame({'$t$': dagen, '$V$': volumes}).T
plt.axhline(0.0, color='k'); plt.axvline(0.0, color='k')
plt.plot(dagen, volumes, 'or')
plt.title('Tumor volume growth'); plt.xlabel('$t$ (days)'); plt.ylabel('$V(t)$ (mmÂ³)')
plt.grid(True); plt.box(False)
plt.show()
```

```python
from math import exp
from ode_solver import ode_solver

n = 20
v = 10
dt = 1.0 / n
ts_exact = [i * dt for i in range(n+1)]
ys_exact = [exp(t) - 1.0 for t in ts_exact]

model = ode_solver(v,n,dt)
x = model.fit(ys_exact)
print(x)
```
